**6. Usa la línea de comandos para descubrir algo interesante sobre los datos dados.**

Descubrimos que se puede contar la cantidad de veces que aparece cada plato en los pedidos con el siguiente comando:

cut -f 3 data.tsv | sort | uniq -c | sort -nr

El resultado será una lista de los platos junto con la cantidad de veces que aparecen en los pedidos, en orden descendente.
